```python
#Renommer les colonnes
df=df.rename(columns={
    'RowNumber':'Numero de ligne',
    'CustomerId': 'ID client',
    'Surname':'Nom de famille',
    'Creditscore':'Score de credit',
    'Geography':'Pays',
    'Gender':'Sexe',
    'Tenure':'AnciennetÃ©(en annÃ©es)',
    'Balance':'Solde',
    'NumOfProducts':'Nombre de produits',
    'HasCrCard':'PossÃ¨de carte de crÃ©dit',
    'IsActiveMember':'Membre actif',
    'EstimatedSalary':'Salaire estimÃ©',
    'Complain': 'Reclamtion faite',
    'Satisfaction Score':'Score de satisfaction',
    'Card Type':'Type de carte bancaire',
    'Point Earned':'Points fidÃ©litÃ© gagnÃ©s',
    'Exited': 'desabonnÃ©'


})
df
```

```python
#Je veux maintenant supprimer les carateres speciaux s'il y'en a dans certaines colonnes
df["Nom de famille"]=df["Nom de famille"].str.replace('[^a-zA-Z0-9]','',regex=True)
df["Pays"]=df["Pays"].str.replace('[^a-zA-Z0-9]','',regex=True)
df["Type de carte bancaire"]=df["Type de carte bancaire"].str.replace('[^a-zA-Z0-9]','',regex=True )
df
```

```python
#Nous voulons calculer le pourcentage du taux 'attrition
taux= df["desabonnÃ©"].mean()*100
print(f"taux {round(taux,2)}%")
```

```python
#Informations sur les clients qui se sont desabonnÃ©
df_des=df[df["desabonnÃ©"] == 1][["Age", "Sexe", "CreditScore", "Pays", "AnciennetÃ©(en annÃ©es)", "Solde"]]
df_des_mean=df_des.select_dtypes(include='number').mean()
print(f"{round(df_des_mean,0)}%")
```

```python
#Pays avec le plus de client desabonnÃ© par pays
churn_par_pays=df.groupby("Pays")["desabonnÃ©"].sum()
bars = plt.bar(churn_par_pays.index, churn_par_pays.values, color='skyblue')
for bar in bars.patches:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, int(yval), ha='center', va='bottom', fontsize=10)
    
plt.title("Nombre de clients dÃ©sabonnÃ©s par pays")
plt.xlabel("Pays")
plt.ylabel("Nombre de dÃ©sabonnÃ©s")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()    
```

```python
client_parti_par_sexe=df[df["desabonnÃ©"]==1].groupby('Sexe')["desabonnÃ©"].count()
client_parti_par_sexe
```

```python
client_fidele_par_sexe= df[df["desabonnÃ©"]==0].groupby('Sexe')["desabonnÃ©"].count()
client_fidele_par_sexe
```

```python
#Nombre de client qui ont fait une reclamation et qui sont parti
# O Pour 'n'a pas fait de reclamation' et 1 pour le contraire 
rec_faite=df.groupby("Reclamtion faite")["desabonnÃ©"].sum()
bars_2=plt.bar(rec_faite.index,rec_faite.values,color='red') 
for bar in bars_2.patches:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, int(yval), ha='center', va='bottom', fontsize=10)

plt.title("Client parti apres reclamation")
plt.xlabel("Pays")
plt.ylabel("Nombre de dÃ©sabonnÃ©s")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()    
```

```python
#Taux de dÃ©sabonnement par score de satisfaction
churn_par_satisfaction = df.groupby("Score de satisfaction")["desabonnÃ©"].mean() * 100
print(f"taux en % {round(churn_par_satisfaction,2)}%")
```

```python


bars_3=churn_par_satisfaction.sort_index().plot(kind="bar", color="orange")

for bar in bars_3.patches:
    yval = bar.get_height()
    bars_3.text(bar.get_x() + bar.get_width()/2, yval + 0.5, f"{yval:.1f}%", 
                ha='center', va='bottom', fontsize=10)
    
plt.title("Taux de dÃ©sabonnement par score de satisfaction")
plt.xlabel("Score de satisfaction (1 = trÃ¨s faible, 5 = trÃ¨s Ã©levÃ©)")
plt.ylabel("Taux de dÃ©sabonnement (%)")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
#Nombre de client fidele par carte bancaire 
fideles_par_carte = df[df["desabonnÃ©"] == 0].groupby("Type de carte bancaire")["desabonnÃ©"].count()
fideles_par_carte
```

```python
#Nombre de client infidele par carte bancaire 
infidele_par_cb=df[df["desabonnÃ©"]==1].groupby("Type de carte bancaire")["desabonnÃ©"].count()
infidele_par_cb
```

```python
taux_fidelite = df.groupby("Type de carte bancaire")["desabonnÃ©"].apply(lambda x: 100 * (1 - x.mean()))
print(f"taux en % {round(taux_fidelite,2)}%")
```

```python
#graphique pour le taux de fidelitÃ© 
bars_4=taux_fidelite.plot(kind='bar', color='seagreen')


for bar in bars_4.patches:   
    yval = bar.get_height()
    bars_4.text(bar.get_x() + bar.get_width()/2, yval + 0.5, f"{yval:.1f}%", 
                ha='center', va='bottom', fontsize=10)
    
taux_fidelite.plot(kind='bar', color='seagreen')
plt.title("Taux de fidÃ©litÃ© par type de carte bancaire")
plt.xlabel("Type de carte bancaire")
plt.ylabel("Taux de fidÃ©litÃ© (%)")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
