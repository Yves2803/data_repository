```python
#Maintenant je vais renommer les colonnes de mon dataset pour que ca me soit plus comprehensible
df
```

```python
#Renommer les colonnes
df=df.rename(columns={
    'RowNumber':'Numero de ligne',
    'CustomerId': 'ID client',
    'Surname':'Nom de famille',
    'Creditscore':'Score de credit',
    'Geography':'Pays',
    'Gender':'Sexe',
    'Tenure':'Ancienneté(en années)',
    'Balance':'Solde',
    'NumOfProducts':'Nombre de produits',
    'HasCrCard':'Possède carte de crédit',
    'IsActiveMember':'Membre actif',
    'EstimatedSalary':'Salaire estimé',
    'Complain': 'Reclamtion faite',
    'Satisfaction Score':'Score de satisfaction',
    'Card Type':'Type de carte bancaire',
    'Point Earned':'Points fidélité gagnés',
    'Exited': 'desabonné'


})
df
```

```python
#Je veux maintenant supprimer les carateres speciaux s'il y'en a dans certaines colonnes
df["Nom de famille"]=df["Nom de famille"].str.replace('[^a-zA-Z0-9]','',regex=True)
df["Pays"]=df["Pays"].str.replace('[^a-zA-Z0-9]','',regex=True)
df["Type de carte bancaire"]=df["Type de carte bancaire"].str.replace('[^a-zA-Z0-9]','',regex=True )
df
```

```python
#Nous voulons calculer le pourcentage du taux 'attrition
taux= df["desabonné"].mean()*100
print(f"taux {round(taux,2)}%")
```

```python
#Informations sur les clients qui se sont desabonné
df_des=df[df["desabonné"] == 1][["Age", "Sexe", "CreditScore", "Pays", "Ancienneté(en années)", "Solde"]]
df_des_mean=df_des.select_dtypes(include='number').mean()
print(f"{round(df_des_mean,0)}%")
```

```python
#Pays avec le plus de client desabonné par pays
churn_par_pays=df.groupby("Pays")["desabonné"].sum()
bars = plt.bar(churn_par_pays.index, churn_par_pays.values, color='skyblue')
for bar in bars.patches:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, int(yval), ha='center', va='bottom', fontsize=10)
    
plt.title("Nombre de clients désabonnés par pays")
plt.xlabel("Pays")
plt.ylabel("Nombre de désabonnés")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()    
```

```python
client_parti_par_sexe=df[df["desabonné"]==1].groupby('Sexe')["desabonné"].count()
client_parti_par_sexe
```

```python
client_fidele_par_sexe= df[df["desabonné"]==0].groupby('Sexe')["desabonné"].count()
client_fidele_par_sexe
```

```python
#Nombre de client qui ont fait une reclamation et qui sont parti
# O Pour 'n'a pas fait de reclamation' et 1 pour le contraire 
rec_faite=df.groupby("Reclamtion faite")["desabonné"].sum()
bars_2=plt.bar(rec_faite.index,rec_faite.values,color='red') 
for bar in bars_2.patches:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.5, int(yval), ha='center', va='bottom', fontsize=10)

plt.title("Client parti apres reclamation")
plt.xlabel("Pays")
plt.ylabel("Nombre de désabonnés")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()    
```

```python
#Taux de désabonnement par score de satisfaction
churn_par_satisfaction = df.groupby("Score de satisfaction")["desabonné"].mean() * 100
print(f"taux en % {round(churn_par_satisfaction,2)}%")
```

```python


bars_3=churn_par_satisfaction.sort_index().plot(kind="bar", color="orange")

for bar in bars_3.patches:
    yval = bar.get_height()
    bars_3.text(bar.get_x() + bar.get_width()/2, yval + 0.5, f"{yval:.1f}%", 
                ha='center', va='bottom', fontsize=10)
    
plt.title("Taux de désabonnement par score de satisfaction")
plt.xlabel("Score de satisfaction (1 = très faible, 5 = très élevé)")
plt.ylabel("Taux de désabonnement (%)")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
#Nombre de client fidele par carte bancaire 
fideles_par_carte = df[df["desabonné"] == 0].groupby("Type de carte bancaire")["desabonné"].count()
fideles_par_carte
```

```python
#Nombre de client infidele par carte bancaire 
infidele_par_cb=df[df["desabonné"]==1].groupby("Type de carte bancaire")["desabonné"].count()
infidele_par_cb
```

```python
taux_fidelite = df.groupby("Type de carte bancaire")["desabonné"].apply(lambda x: 100 * (1 - x.mean()))
print(f"taux en % {round(taux_fidelite,2)}%")
```

```python
#graphique pour le taux de fidelité 
bars_4=taux_fidelite.plot(kind='bar', color='seagreen')


for bar in bars_4.patches:   
    yval = bar.get_height()
    bars_4.text(bar.get_x() + bar.get_width()/2, yval + 0.5, f"{yval:.1f}%", 
                ha='center', va='bottom', fontsize=10)
    
taux_fidelite.plot(kind='bar', color='seagreen')
plt.title("Taux de fidélité par type de carte bancaire")
plt.xlabel("Type de carte bancaire")
plt.ylabel("Taux de fidélité (%)")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
